# apply functions

```python
def rating_function(x):
    if x >= 8.0:
        return "good"
    else:
        return "bad"
```

```python
movies_df["rating_category"] = movies_df["rating"].apply(rating_function)

movies_df.head(2)
```

```python
movies_df["rating_category"] = movies_df["rating"].apply(lambda x: 'good' if x >= 8.0 else 'bad')

movies_df.head(2)
```

# pandas aggregation

```python
movies_df['new_genre'] = movies_df['genre'].apply(lambda x:x.split(',')[0])
```

```python
movies_df.head()
```

```python
# movies_df = movies_df.rename(columns={'Revenue (Millions)':'Revenue_millions'})
```

```python
movies_df_gb = movies_df[['year','new_genre','revenue_millions']].groupby(['year','new_genre']).agg({'new_genre':'count','revenue_millions':'sum'})
# select year, new_genre, count(new_genre), sum(revenue_millions) 
# group by year, new_genre
movies_df_gb.columns = ['count_genre','sum_revenue_mio']
movies_df_gb = movies_df_gb.reset_index()
```

```python
movies_df_gb.head()
```

```python
#pivot
movies_df_gb_pvt = pd.pivot_table(movies_df_gb, values='count_genre', 
                                  index=['year'],columns=['new_genre'], aggfunc=np.sum).fillna(0)
```

```python
movies_df_gb_pvt.head()
```

# Join

```python
raw_data = {
        'subject_id': ['1', '2', '3', '4', '5'],
        'first_name': ['Alex', 'Amy', 'Allen', 'Alice', 'Ayoung'], 
        'last_name': ['Anderson', 'Ackerman', 'Ali', 'Aoni', 'Atiches']}
df_a = pd.DataFrame(raw_data, columns = ['subject_id', 'first_name', 'last_name'])
raw_data = {
        'subject_id': ['4', '5', '6', '7', '8'],
        'first_name': ['Billy', 'Brian', 'Bran', 'Bryce', 'Betty'], 
        'last_name': ['Bonder', 'Black', 'Balwner', 'Brice', 'Btisan']}
df_b = pd.DataFrame(raw_data, columns = ['subject_id', 'first_name', 'last_name'])
raw_data = {
        'subject_id': ['1', '2', '3', '4', '5', '7', '8', '9', '10', '11'],
        'test_id': [51, 15, 15, 61, 16, 14, 15, 1, 61, 16]}
df_n = pd.DataFrame(raw_data, columns = ['subject_id','test_id'])
```

```python
#union
df_new = pd.concat([df_a, df_b])
```

```python
df_new
```

```python
#join without key
df_new2 = pd.concat([df_a, df_b], axis=1)
```

```python
#join with key
df_new3 = pd.merge(df_new, df_n, on='subject_id')
```

```python
df_new3
```

```python
#join with key 
df_new4 = pd.merge(df_new, df_n, left_on='subject_id', right_on='subject_id')
```

```python
#inner join 
df_new5 = pd.merge(df_a, df_b, on='subject_id', how='inner')
```

```python
# alternative aggregation method
df_alt = df['data'].reset_index()
df_alt['year'] = df_alt['datetime'].apply(lambda x:x.year)
df_alt['month'] = df_alt['datetime'].apply(lambda x:x.year*100 + x.month)
df_alt['day'] = df_alt['datetime'].apply(lambda x:x.year*10000 + x.month*1000 + x.day)
df_alt['hour'] = df_alt['datetime'].apply(lambda x:x.hour)
df_alt['day_name'] = df_alt['datetime'].apply(lambda x:x.day_name())
```

```python
df_alt
```

```python
# summary by hour 
df_alt_h_mean = df_alt[['hour','data']].groupby('hour').agg({'data':'mean'}) 
df_alt_h_d_mean = df_alt[['day_name','hour','data']].groupby(['day_name','hour']).agg({'data':'mean'}) 
```

```python
#pivot day_name and hour 
df_alt_h_d_mean_pvt = pd.pivot_table(df_alt_h_d_mean, values='data', index=['day_name'],columns=['hour'], aggfunc=np.sum).fillna(0)
```
